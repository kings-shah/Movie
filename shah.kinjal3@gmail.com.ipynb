```python
def gender_fun(g):
    if g in adventure_action:
        return 'Adventure & Action'
    elif g in romance:
        return 'Romance'
    elif g in thriller:
        return 'Thriller & Mystry'
    elif g in horror:
        return 'Horror'
    elif g in drama:
        return 'Drama'
    elif g in comedy:
        return 'Comedy'
    elif g in sci_fant:
        return 'Sci-fi & Fantasy'
data['genre_new']=data['genre'].apply(gender_fun)
```

###### Handling null values in Revenue colmn.
We oserved that revenue col is affected by the rating. Movies with rating aove 6 have more revenue than movies with rating below 6. So we will replace the null values with median of revene according to the genre

```python
for (row,col) in data.iterrows():
    if (col.genre_new=='Drama')& (col.rating>=6.0) & (pd.isnull(col.revenue)):
        data.loc[row,'revenue']=15.96
    if (col.genre_new=='Drama')& (col.rating<6.0) & (pd.isnull(col.revenue)):
        data.loc[row,'revenue']=25.98
    if (col.genre_new=='Horror')& (col.rating>=6.0) & (pd.isnull(col.revenue)):
        data.loc[row,'revenue']=29.85
    if (col.genre_new=='Horror')& (col.rating<6.0) & (pd.isnull(col.revenue)):
        data.loc[row,'revenue']=25.92
    if (col.genre_new=='Thriller & Mystry')& (col.rating>=6.0) & (pd.isnull(col.revenue)):
        data.loc[row,'revenue']=32.07
    if (col.genre_new=='Thriller & Mystry')& (col.rating<6.0) & (pd.isnull(col.revenue)):
        data.loc[row,'revenue']=26.18
    if (col.genre_new=='Sci-fi & Fantasy')& (col.rating>=6.0) & (pd.isnull(col.revenue)):
        data.loc[row,'revenue']=100.98
    if (col.genre_new=='Sci-fi & Fantasy')& (col.rating<6.0) & (pd.isnull(col.revenue)):
        data.loc[row,'revenue']=63.40
    if (col.genre_new=='Adventure & Action')& (col.rating>=6.0) & (pd.isnull(col.revenue)):
        data.loc[row,'revenue']=85.95
    if (col.genre_new=='Adventure & Action')& (col.rating<6.0) & (pd.isnull(col.revenue)):
        data.loc[row,'revenue']=34.01
    if (col.genre_new=='Comedy')& (col.rating>=6.0) & (pd.isnull(col.revenue)):
        data.loc[row,'revenue']=42.82
    if (col.genre_new=='Comedy')& (col.rating<6.0) & (pd.isnull(col.revenue)):
        data.loc[row,'revenue']=48.24
    if (col.genre_new=='Romance')& (col.rating>=6.0) & (pd.isnull(col.revenue)):
        data.loc[row,'revenue']=42.82
    if (col.genre_new=='Romance')& (col.rating<6.0) & (pd.isnull(col.revenue)):
        data.loc[row,'revenue']=44.67
```

###### Creating 4 Actors columns
As observed, Each row in the actor column has 4 actor names comma separated 

```python
data['actor1']=None
data['actor2']=None
data['actor3']=None
data['actor4']=None
for (row,col) in data.iterrows():
    l1=col.actors.split(',')
    l1.sort()
    i=0
    if len(l1)>0:
        data.loc[row,'actor1']=l1[0].strip().casefold()
    if len(l1)>1:
        data.loc[row,'actor2']=l1[1].strip().casefold()
    if len(l1)>2:
        data.loc[row,'actor3']=l1[2].strip().casefold()
    if len(l1)>3:
        data.loc[row,'actor4']=l1[3].strip().casefold()
```

```python
P = pd.DataFrame(columns = ["Name","Count","Total_Revenue"])
for A_names in S:
    sets = [A_names]
    m = data.loc[((data['genre_new']=='Sci-fi & Fantasy')|(data['genre_new']=='Adventure & Action'))&((data['actor1'].str.replace('s/+',"")==(A_names)) | (data['actor4'].str.replace('s/+',"")==(A_names)) |(data['actor3'].str.replace('s/+',"")==(A_names)) |(data['actor2'].str.replace('s/+',"")==(A_names)))].agg({"year": "count" , "revenue": "sum"})
    # if the actor name matches and the movie belongs to Sci-fi & Fantasy or Adventure & Action genre then we sum the revenue and inc the count
    sets.append(m[0])
    sets.append(m[1])
    a_series = pd.Series(sets, index = P.columns) #putting the count, sum of revenue, actor name to series
    P = P.append(a_series, ignore_index = True)
P["Avg_Rev"] = P.Total_Revenue/P.Count 
P.sort_values(by = 'Total_Revenue', ascending = False,inplace = True)
P = P.head(10).sort_values(ascending=False,by='Avg_Rev') # storing the top 10 actor names
```
